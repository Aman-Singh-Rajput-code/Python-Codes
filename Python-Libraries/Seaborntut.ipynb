```python
#Introduction of Seaborn:
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

var = [1,2,3,4,5,6,7]
var1 = [2,3,4,1,5,6,7]

x1 = pd.DataFrame({"var":var,"var1":var1})

sns.lineplot(x = "var",y = "var1",data = x1)  #to directly use dataset without dataframe (x = var) pass argument like this
plt.show()

plt.plot(var,var1)
plt.show()
```

```python
#Working on big data set
#part1
y = sns.load_dataset("penguins")  #this is used to load entire dataset
y1 = sns.load_dataset("penguins").head(20)  #'''this is used to load specific dataset(subset)'''
y1  #used to show subset data
y   #used to show entire data

#part2
sns.lineplot(x = "bill_length_mm",y = "flipper_length_mm",data = y1,hue = "sex",size=10 )  #similar to legend in matpotlib
plt.show()

#part3
sns.lineplot(x = "bill_length_mm",y = "flipper_length_mm",data = y1,hue = "sex",size=10,style="sex",palette="Accent",markers=[">","o"],dashes=False,legend=False)
                                                                   #legend     #size   #Design      #diff.colors    #for plots marks   #used to dASH   #remove legend
plt.grid()                                                                                                                           #line need or not  #or add diff legend  
plt.title("Bird Data")
plt.show()
```

Histogram in seaborn


```python
#part1
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
var = sns.load_dataset("penguins")
var
```

```python
#part2 
sns.displot(var["bill_length_mm"],kde=True,rug=True)  #kde(kernal dencity estimation)
plt.show()                                 #rug is used to make trace lines for KDE

sns.displot(var["flipper_length_mm"],bins=[170,180,190,200,210,220,230,240])   #bins is used to changes and modify interval
plt.show()

sns.displot(var["bill_depth_mm"],color="y",log_scale=True) #to convert all data into logs
plt.show()
```

Barplot in Seaborn

```python
#part1
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
var = sns.load_dataset("penguins")
var
```

```python
#part2
sns.barplot(x=var.island,y = var.bill_length_mm) #both are equal (x="island",y="bill_length_mm",data=var)
plt.show()
```

```python
#part3
sns.barplot(x="island",y="bill_length_mm",data=var,hue="sex",errorbar=('ci', 50)) #hue is used to divide bars as per condition like"sex"
plt.show()                                                    #ci is used to handle barlines  #dodge is used to overlap bars
```

```python
#Part4
sns.set(style="darkgrid")  #used for grid in sns
order1 = ["Dream","Torgersen","Biscoe"]
sns.barplot(x="island",y="bill_length_mm",data=var,hue="sex",
           order = order1,hue_order = ["Female","Male"],palette="icefire",alpha=0.5) #order is used to manupulate order of bars 
plt.show()                                                                #it is used to light the colors
```

```python
#part5
order1 = ["Dream","Torgersen","Biscoe"]
sns.barplot(x="island",y="bill_length_mm",data=var,color = "blue",saturation=0.1,err_kws={'color': 'red','linewidth':1},capsize = 0.5) #saturation is used to balance the color
plt.show()                                                                        #this is used to handle barline colors,width,etc    #cap size is for bars upper&lower lines
```

Scatter plot in Seaborn

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
data = sns.load_dataset("penguins")
data
m={"Male":"*","Female":"o"}
sns.scatterplot(x = "bill_length_mm",y = "bill_depth_mm",data=data,hue="sex",style="sex",
               size="sex",sizes=(100,40),palette="gist_ncar",alpha=0.5,markers=m)
plt.show()
```
