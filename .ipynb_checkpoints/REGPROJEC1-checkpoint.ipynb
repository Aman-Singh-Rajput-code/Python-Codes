```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
#Display all the columns of Dataframe

pd.pandas.set_option('display.max_columns',None)
```

```python
PROJECT_DIR = r"C:\Users\Aman\OneDrive\Documents\Feature-engineering-project"
DATA_DIR = "Data"
```

```python
def get_data(name):
    fname = f"{name}.csv"
    fpath = os.path.join(PROJECT_DIR,DATA_DIR,fname)
    return pd.read_csv(fpath,usecols=['YearBuilt','SalePrice'])
```

```python
df = get_data('train')
```

```python
df.head()
```

```python
plt.scatter(df['YearBuilt'],df['SalePrice'])
plt.xlabel('YearBuilt')
plt.ylabel('Saleprice(in lpa)')
plt.show()
```

```python
X = df.iloc[:,0:1]
y = df.iloc[:,-1]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
X_test
```

```python
y_test
```

```python
lr.predict(X_test.iloc[0].values.reshape(1,1))
```

```python
plt.scatter(df['YearBuilt'],df['SalePrice'],color='yellow')
plt.plot(X_train,lr.predict(X_train),color='red')
plt.xlabel('Year')
plt.ylabel('SalePrice(in lpa)')
plt.show()
```

```python
m = lr.coef_
```

```python
b = lr.intercept_
```
